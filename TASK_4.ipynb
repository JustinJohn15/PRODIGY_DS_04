```python
import pandas as pd

file_path = "/content/twitter_validation.csv"
data = pd.read_csv(file_path)

print(data.head())
```

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def calculate_sentiment(text):
    analysis = TextBlob(str(text))
    return analysis.sentiment.polarity

data['Sentiment'] = data['Irrelevant'].apply(calculate_sentiment)
```

```python
sentiment_over_time = data.groupby('Facebook')['Sentiment'].mean()
print(sentiment_over_time)
```

```python
import matplotlib.pyplot as plt

# Plot sentiment over time
plt.figure(figsize=(10, 6))
sentiment_over_time.plot()
plt.title('Sentiment Over Time')
plt.xlabel('Facebook')
plt.ylabel('Average Sentiment')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
